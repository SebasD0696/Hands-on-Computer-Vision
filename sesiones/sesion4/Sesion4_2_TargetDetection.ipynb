# <font color='#4C5FDA'>**Objetivo**</font>


El objetivo de la detecci칩n de objetivos hiperespectrales es detectar si un objetivo concreto se encuentra en cada p칤xel de un cubo de datos hiperespectrales.  En general, los enfoques de detecci칩n de objetivos no son eficaces para estimar *cu치nto* de un objetivo se encuentra dentro del campo de visi칩n de un p칤xel (eso ser칤a desmezclar) - s칩lo si el material est치 all칤 o no (con cierta confianza).  

Los algoritmos de detecci칩n de objetivos hiperespectrales m치s utilizados son el *Spectral Matched Filter* (SMF) y el *Adaptive Cosine Estimator* (ACE).  Ambos m칠todos requieren tener de antemano la firma del blanco que se busca.  Ambos m칠todos estiman la confianza en el objetivo calculando el coseno del 치ngulo entre la firma del objetivo y un p칤xel de prueba en un espacio blanqueado (es decir, tras la normalizaci칩n del fondo).  La principal diferencia entre ambos m칠todos es que el SMF tiene en cuenta la magnitud, mientras que el ACE no.  Se calculan del siguiente modo:


$SMF = \frac{\left(\left(\mathbf{s}-\mu\right)^T \Sigma^{-1}\left(\mathbf{x}-\mu\right)\right)^2}{\left(\mathbf{s}-\mu\right)^T \Sigma^{-1}\left(\mathbf{s}-\mu\right)}$

$ACE = \frac{\left(\left(\mathbf{s}-\mu\right)^T \Sigma^{-1}\left(\mathbf{x}-\mu\right)\right)^2}{\left(\mathbf{s}-\mu\right)^T \Sigma^{-1}\left(\mathbf{s}-\mu\right)\left(\mathbf{x}-\mu\right)^T \Sigma^{-1}\left(\mathbf{x}-\mu\right)}$

donde $\mathbf{s}$ es la firma del objetivo, $\mathbf{x}$ es el p칤xel bajo prueba, $\mu$ es la media del fondo, y $\Sigma$ es la covarianza del fondo.

<font color='#4C5FDA'>**Estimaci칩n de la media y la covarianza del fondo**</font>

A menudo resulta dif칤cil determinar c칩mo estimar la media y la covarianza del fondo.  Si hay muy pocos objetivos en una escena, puede bastar con utilizar la media y la covarianza globales de la escena para representar el fondo.  Tambi칠n se ha trabajado para segmentar la escena y, a continuaci칩n, calcular las estad칤sticas del fondo para cada segmento.

Tambi칠n hay enfoques en los que las estad칤sticas de fondo se calculan en un enfoque de "ventana deslizante" para centrarse s칩lo en la informaci칩n local durante la estimaci칩n del fondo. El problema de estos m칠todos es determinar el tama침o de la ventana.  Si la ventana es demasiado peque침a, puede tener problemas para estimar las estad칤sticas de fondo de forma fiable.  Sin embargo, si la ventana es demasiado grande, ya no se utiliza s칩lo la informaci칩n local.

# <font color='#4C5FDA'>**Importamos librerias**</font>

```python
# Importaciones y configuraci칩n
import numpy as np
import os.path
import scipy.io
from loadmat import loadmat

import matplotlib as mpl
%matplotlib inline
default_dpi = mpl.rcParamsDefault['figure.dpi']
mpl.rcParams['figure.dpi'] = default_dpi*2
import matplotlib.pyplot as plt

from hsi_detectors import smf_detector,ace_detector
```

```python
# Cargar la imagen del Campus
img_fname = 'muufl_gulfport_campus_w_lidar_1.mat'
spectra_fname = 'tgt_img_spectra.mat'

dataset = loadmat(img_fname)['hsi']

hsi = dataset['Data']
n_r,n_c,n_b = hsi.shape
wvl = dataset['info']['wavelength']
rgb = dataset['RGB']
```

```python
# Carga el conjunto de datos de espectros desde un archivo MAT
spectra_dataset = loadmat(spectra_fname)
# Extrae los espectros de los objetivos (targets) del conjunto de datos
tgts = spectra_dataset['tgt_img_spectra']['spectra']
# Extrae los nombres de los objetivos (targets) del conjunto de datos
tgt_names = spectra_dataset['tgt_img_spectra']['names']
```

```python
# Mira los nombres de los objetivos
tgt_names
```

```python
# NO MODIFIQUES ESTA CELDA
# Trazar los espectros de la firma del objetivo y graficarlos

# Recorta los espectros de los objetivos para que coincidan con el n칰mero de longitudes de onda disponibles en 'wvl'
trimmed_tgts = tgts[:wvl.shape[0], :]

# Itera sobre cada espectro de objetivo en el conjunto de datos recortado
for i in range(trimmed_tgts.shape[1]):
    # Grafica el espectro del objetivo i contra las longitudes de onda
    plt.plot(wvl, trimmed_tgts[:, i], label=tgt_names[i])

plt.legend()
plt.ylim(0, 1)
plt.ylabel('Reflectance (%)')
plt.xlabel('Wavelength (nm)')
plt.title('MUUFL Target Signatures')
plt.show()
```
